```python
import os
import numpy as np
import cv2 
# from sklearn import svm
import shutil
import scipy.io
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# from imutils import face_utils
# import argparse
# import imutils
# import dlib

def insertion(arr,i):
    if(float(arr[i]) > 0):
        return 1
    else:
        return 0
```

```python
x = np.loadtxt('full_attributes',delimiter = '\n',dtype=np.str)
print(len(x))
array = []
h, w = 13143, 74
matrix = [[0 for x in range(w)] for y in range(h)] 
images = []
#---------------------
# we have 74 attributes - and 2 for the name of the person
for i in range(len(x)):
    arr = x[i].split() # 76 = len(arr)
    
    name_len = len(arr)-74
# Adding name to matrix
    la = arr[0]+"_"
    for l in range(1,name_len):
        la += arr[l]+"_"
    if(float(arr[name_len]) > 99):
        la += "0" + arr[name_len]+".ppm"
    if(float(arr[name_len]) > 9 and float(arr[name_len]) < 100):
        la += "00" + arr[name_len] +".ppm"    
    if(float(arr[name_len]) < 10):
        la += "000" + arr[name_len] +".ppm"
    matrix[i][0] = la
    images.append(la)
    
    for j in range(name_len,len(arr)-1):
        if(float(arr[j+1]) > 0): matrix[i][j-name_len+1] = 1 
        else: matrix[i][j-name_len+1] = 0
        insertion(arr,j+1)
print("Matrix size",len(matrix),len(matrix[0]))
```
